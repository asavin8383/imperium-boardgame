```python
import os
from os import listdir
from os.path import isfile, join
import re
import pandas as pd
import datetime


from mfo_ner.prediction.predictor import Predictor

now = datetime.datetime.now()
predictor = Predictor()
# with open("../../test_results/social_results.csv", 'w') as output_file:
df = pd.DataFrame(columns = ['Message', 'NER'])
# csv_writer = csv.writer(output_file, dialect='excel', lineterminator='\n')
dirs = [x[0] for x in os.walk("../../test_files/social")]
for cur_dir in dirs:
    files = [f for f in listdir(cur_dir) if isfile(join(cur_dir, f))]
    for file in files:
        try:
            lines = open(cur_dir + "/" + file, "r", encoding='utf-8').readlines()
            for line in lines:
                line = line.strip().replace("\t", " ").replace("\n", " ").replace("\r", "")
                line = re.sub(r"^\"+", "", line)
                line = re.sub(r"\"+$", "", line)
                entities = predictor.predict_by_html(line)
                entities_str = ""
                df_loc = pd.DataFrame([[line, entities]], columns = ['Message', 'NER'])
                df = df.append(df_loc)
            df.to_excel("../../test_results/social_results"+str(now.microsecond)+".xlsx")
        except Exception as ex:
            print(ex)
            print("Error file: " + file)
        print(file + " completed")
print("completed")
```
